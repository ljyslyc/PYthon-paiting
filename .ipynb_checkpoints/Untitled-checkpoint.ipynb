```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt #引入库
R = 0.02
ma = 0.1
mb = 0.1
k = 20000
df = 1.0
time = 0.1
dt = 0.0001 #设置数值

n = int(time/dt)
t = np.zeros(n,float)
ra = np.zeros((n,2),float)
rb = np.zeros((n,2),float)
va = np.zeros((n,2),float)
vb = np.zeros((n,2),float)
f = np.zeros((n,2),float)  #设置数值

ra[0] = array([0.0,0.0])
rb[0] = array([0.08,0.0])
va[0] = array([1.0,0.0])
vb[0] = array([0.0,0.0])  #初始化
for i in range(n-1):
    dr = sqrt((ra[i][0] - rb[i][0])**2+(ra[i][1]-rb[i][1])**2)
    if dr >= 2*R:
        f[i] = 0 
        va[i+1] = va[i]
        vb[i+1] = vb[i] #r>=2R时
    else:
        f[i] = -k*(abs(dr - 2*R))**(3/2)*(rb[i]-ra[i])/dr + df*(vb[i]-va[i])
        va[i+1] = f[i]*dt/ma + va[i]
        vb[i+1] = -f[i]*dt/mb + vb[i] #r<2R时
    ra[i+1] = ra[i] + va[i+1]*dt
    rb[i+1] = rb[i] + vb[i+1]*dt
    t[i+1] = t[i] + dt  
th = linspace(-pi,pi,100)
plt.plot(ra[:,0],ra[:,1],"-b",R*cos(th),R*sin(th),"-b",rb[:,0],rb[:,1],"-r",0.08+R*cos(th),R*sin(th),"-r")
plt.axis([-0.02,0.12,-0.04,0.06])
plt.show()  #画两球在平面上的轨迹
plt.subplot(211)
plt.plot(t,sqrt(va[:,0]**2+va[:,1]**2))
plt.xlabel("t[s]")
plt.ylabel("va[m/s]")
plt.subplot(212)
plt.plot(t,sqrt(vb[:,0]**2+vb[:,1]**2))
plt.xlabel("t[s]")
plt.ylabel("vb[m/s]")
plt.show()  #画出两球的速度随时间的变化
```
